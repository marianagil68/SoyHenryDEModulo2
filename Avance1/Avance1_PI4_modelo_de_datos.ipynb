# Identificación de Claves Primarias, Foráneas y Atributos de Negocio

## DER con las claves foráneas
![DER ecommerce](image.png)

## 1. Tabla: `usuarios`
- **PK:** `id`
- **FKs implícitas:** ordenes.usuario_id, direcciones_envio.usuario_id, resenas_productos.usuario_id, carrito.usuario_id
- **Atributos principales:** dni, email, nombre, apellido
- **Variables de negocio:** identificación, contacto y autenticación del cliente

## 2. Tabla: `categorias`
- **PK:** `id`
- **FKs implícitas:** productos.categoria_id
- **Atributos principales:** nombre, descripcion
- **Variables de negocio:** segmentación de productos

## 3. Tabla: `productos`
- **PK:** `id`
- **FK:** categoria_id → categorias.id
- **Atributos principales:** nombre, descripcion, precio, stock
- **Variables de negocio:** oferta, precio y disponibilidad

## 4. Tabla: `ordenes`
- **PK:** `id`
- **FK:** usuario_id → usuarios.id
- **Atributos principales:** usuario_id, fecha_orden, total, estado
- **Variables de negocio:** ingresos, actividad temporal, ciclo de venta

## 5. Tabla: `detalle_ordenes`
- **PK:** `id`
- **FKs:** orden_id → ordenes.id, producto_id → productos.id
- **Atributos principales:** cantidad, precio_unitario
- **Variables de negocio:** mix de productos, demanda, ingresos detallados

## 6. Tabla: `direcciones_envio`
- **PK:** `id`
- **FK:** usuario_id → usuarios.id
- **Atributos principales:** dirección completa
- **Variables de negocio:** logística y zonas de envío

## 7. Tabla: `carrito`
- **PK:** `id`
- **FKs:** usuario_id → usuarios.id, producto_id → productos.id
- **Atributos principales:** cantidad, fecha_agregado
- **Variables de negocio:** intención de compra y comportamiento del usuario

## 8. Tabla: `metodos_pago`
- **PK:** `id`
- **Atributos principales:** nombre, descripcion
- **Variables de negocio:** mix de métodos de pago

## 9. Tabla: `ordenes_metodos_pago`
- **PK:** `id`
- **FKs:** orden_id → ordenes.id, metodo_pago_id → metodos_pago.id
- **Atributos principales:** monto_pagado
- **Variables de negocio:** distribución de pagos y análisis por método

## 10. Tabla: `resenas_productos`
- **PK:** `id`
- **FKs:** usuario_id → usuarios.id, producto_id → productos.id
- **Atributos principales:** calificacion, comentario, fecha
- **Variables de negocio:** satisfacción del cliente y feedback

## 11. Tabla: `historial_pagos`
- **PK:** `id`
- **FKs:** orden_id → ordenes.id, metodo_pago_id → metodos_pago.id (nullable)
- **Atributos principales:** monto, fecha_pago, estado_pago
- **Variables de negocio:** cobranza, mora y flujo financiero
